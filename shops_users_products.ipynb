```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import feather
from datetime import datetime
import collections
import operator
import xlsxwriter
%matplotlib inline
sns.set_style("ticks")
sns.set_context("notebook", font_scale=1.1)
pd.options.display.max_colwidth = 1000 # afficher max 1000 caract√®res
```

```python
# load the table of products whose prices are already converted into euro
# assign a sufix "products" for all columns
euro_products= feather.read_dataframe('../integration/euro_products.feather')
euro_products.columns = [str(col) + '_products' for col in euro_products.columns]
euro_products
```

```python
# load the table of shops
# assign a sufix "shops" for all columns
shops = pd.ExcelFile('../_outputs/shop_details.xlsx').parse('Sheet1')
shops.columns = [str(col) + '_shops' for col in shops.columns]
shops = shops.rename(columns={'shop_name_shops':'shop_id_products','user_id_shops':'user_id_users'})
shops
```

```python
# load the table of users 
# assign a sufix "users" for all columns
users = pd.ExcelFile('../_outputs/user_profiles.xlsx').parse('Sheet1')
users.columns = [str(col) + '_users' for col in users.columns]
```

```python
# merge products and shops tables
merge1 = euro_products.merge(shops,on='shop_id_products')
```

```python
# merge all tables (products, shops and users)
merge2 = merge1.merge(users,on='user_id_users')
```

```python
# Calculate and add the "mean_price" column
merge2['euro_price_products'] = merge2['euro_price_products'].astype(float)
mean_price = merge2.groupby(by = ['shop_id_products'])['euro_price_products'].mean().map(lambda x:round(x,2)) 
df_mean_price = pd.DataFrame([mean_price]).T.reset_index(level='shop_id_products').rename(columns={'euro_price_products':'mean_price'})
df_mean_price
```

```python
# Calculate and add the "median_price" column
median_price = merge2.groupby(by = ['shop_id_products'])['euro_price_products'].median().map(lambda x:round(x,2))
df_median_price = pd.DataFrame([median_price]).T.reset_index(level='shop_id_products').rename(columns={'euro_price_products':'median_price'})
df_median_price
```

```python
# Calculate and add the "min_price" column
min_price = merge2.groupby(by = ['shop_id_products'])['euro_price_products'].min()
df_min_price = pd.DataFrame([min_price]).T.reset_index(level='shop_id_products').rename(columns={'euro_price_products':'min_price'})
df_min_price
```

```python
# Calculate and add the "max_price" column
max_price = merge2.groupby(by = ['shop_id_products'])['euro_price_products'].max()
df_max_price = pd.DataFrame([max_price]).T.reset_index(level='shop_id_products').rename(columns={'euro_price_products':'max_price'})
df_max_price
```

```python
# Calculate and add the "variance_price" column
merge2['euro_price_products'] = merge2['euro_price_products'].astype(float)
variance_price = merge2.groupby(by = ['shop_id_products'])['euro_price_products'].var().map(lambda x:round(x,2))
df_variance_price = pd.DataFrame([variance_price]).T.reset_index(level='shop_id_products').rename(columns={'euro_price_products':'variance_price'})
df_variance_price
```

```python
# Calculate and add the "mean_views_products" column
mean_views = merge2.groupby(by = ['shop_id_products'])['views_products'].mean().map(lambda x:round(x,2))
df_mean_views = pd.DataFrame([mean_views]).T.reset_index(level='shop_id_products').rename(columns={'views_products':'mean_views_products'})
df_mean_views
```

```python
# Calculate and add the "mean_tags_products" column
merge2['count_list_tags'] = merge2['tags_products'].map(lambda x:len(x.split(',')))
mean_tags = merge2.groupby(by = ['shop_id_products'])['count_list_tags'].mean().map(lambda x:round(x,2))
df_mean_tags = pd.DataFrame([mean_tags]).T.reset_index(level='shop_id_products').rename(columns={'count_list_tags':'mean_tags_products'})
df_mean_tags
```

```python
# Calculate and add the "num_i_did_who_made_products" column
count_who_made_products = merge2.groupby(['shop_id_products', 'who_made_products']).size().reset_index(name='num_i_did_who_made_products')
count_i_did_who_made_products = count_who_made_products.loc[(count_who_made_products['who_made_products']=='i_did')]
num_i_did_who_made_products = count_i_did_who_made_products [['shop_id_products','num_i_did_who_made_products']]
num_i_did_who_made_products

```

```python
# Calculate and add the "num_Someone_else_who_made_products" column
count_who_made_products = merge2.groupby(['shop_id_products', 'who_made_products']).size().reset_index(name='num_Someone_else_who_made_products')
count_Someone_else_who_made_products = count_who_made_products.loc[(count_who_made_products['who_made_products']=='someone_else')]
num_Someone_else_who_made_products = count_Someone_else_who_made_products [['shop_id_products','num_Someone_else_who_made_products']]
num_Someone_else_who_made_products
```

```python
# Calculate and add the "num_collective_who_made_products" column
count_who_made_products = merge2.groupby(['shop_id_products', 'who_made_products']).size().reset_index(name='num_collective_who_made_products')
count_collective_who_made_products = count_who_made_products.loc[(count_who_made_products['who_made_products']=='collective')]
num_collective_who_made_products = count_collective_who_made_products [['shop_id_products','num_collective_who_made_products']]
num_collective_who_made_products
```

```python
# Calculate and add the "mean_favorers_products_per_shop" column
mean_favorers_products_per_shop = merge2.groupby(by =['shop_id_products'])['num_favorers_products'].mean().map(lambda x:round(x,2))
df_mean_favorers_products_per_shop = pd.DataFrame([mean_favorers_products_per_shop]).T.reset_index(level='shop_id_products').rename(columns={'num_favorers_products':'mean_favorers_products_per_shop'})
df_mean_favorers_products_per_shop
```

```python
# Calculate and add the "num_when_made_to_order" column
count_when_made = merge2.groupby(['shop_id_products', 'when_made_products']).size().reset_index(name='num_when_made_to_order')
count_when_made_to_order = count_when_made.loc[(count_when_made['when_made_products']=='made_to_order')]
num_when_made_to_order = count_when_made_to_order[['shop_id_products','num_when_made_to_order']]
num_when_made_to_order
```

```python
# Calculate and add the "num_True_is_customizable_products" column
count_is_customizable_products = merge2.groupby(['shop_id_products', 'is_customizable_products']).size().reset_index(name='num_True_is_customizable_products')
count_True_is_customizable_products = count_is_customizable_products.loc[(count_is_customizable_products['is_customizable_products']=='True')]
num_True_is_customizable_products = count_True_is_customizable_products[['shop_id_products','num_True_is_customizable_products']]
num_True_is_customizable_products
```

```python
# Calculate and add the "num_True_is_digital_products" column
count_is_digital_products = merge2.groupby(['shop_id_products', 'is_digital_products']).size().reset_index(name='num_True_is_digital_products')
count_True_is_digital_products = count_is_digital_products.loc[(count_is_digital_products['is_digital_products']=='True')]
num_True_is_digital_products = count_True_is_digital_products[['shop_id_products','num_True_is_digital_products']]
num_True_is_digital_products
```

```python
# Calculate and add the "3_top_category" column
l = list(merge2['taxonomy_id_products'].groupby(merge2['shop_id_products']))

k = []
for a in range(len(l)-1):
    counter = collections.Counter(l[a][1])
    sorted_counter = sorted(counter, reverse=True)[:3]
    k.append([l[a][0],sorted_counter])

df_3top_category = pd.DataFrame(k,columns = ['shop_id_products','3_top_category'] )
df_3top_category
```

```python
# merge users and shops tables
shops_users= users.merge(shops,on='user_id_users')
```

```python
import functools

# Reduce function with three parameters
 def reduce(function, iterable, initializer=None):
    it = iter(iterable)
    if initializer is None:
        value = next(it)
    else:
        value = initializer
    for element in it:
        value = function(value, element)
    return value

# list of all new columns already calculated    
data_frames = [shops_users,df_mean_tags,df_mean_price,df_median_price,df_variance_price,df_min_price,df_max_price,
               df_mean_views,df_mean_favorers_products_per_shop,df_3top_category,num_i_did_who_made_products,
               num_collective_who_made_products,num_Someone_else_who_made_products,num_when_made_to_order,
               num_True_is_customizable_products,num_True_is_digital_products] 

# merge shops_users table with all new columns
df_calculs = reduce(lambda  left,right: pd.merge(left,right,on=['shop_id_products'],how='outer'), data_frames)
df_calculs
```

```python
# calculate and add the age of the shops from "creation_date_shops" to 2017/9/15

df_calculs['creation_date_shops'] = pd.to_datetime(df_calculs['creation_date_shops'],infer_datetime_format=True)

from datetime import *

def months_between(d1, d2):
    y=abs(d1.year - d2.year)
    m=abs(d1.month - d2.month)
    return (str(y)+' years and '+str(m)+' months')

df_calculs['Age_shops'] = df_calculs['creation_date_shops'].apply(lambda d : months_between(date(2017,9,15),d))
df_calculs[['shop_id_products','Age_shops','creation_date_shops']]
```

```python
# calculate and add "CA" column 
df_calculs['CA'] = df_calculs.transaction_sold_count_users * df_calculs.mean_price
```
